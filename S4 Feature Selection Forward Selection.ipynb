## Preprocessing for feature selection

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_cat = X[cat]
X_cat_enc = X_cat.apply(le.fit_transform)
```

```python
X_cat_enc
```

## Standard Scaling of continuous features

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_con = X[con]
X_con_s = pd.DataFrame(scaler.fit_transform(X_con),columns=con)
```

```python
X_con.head()
```

## Joining Categorical and Continuous features

```python
Xnew = X_con_s.join(X_cat_enc)
```

```python
Xnew.head()
```

## Start feature Selection

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
selector = SequentialFeatureSelector(model,direction='forward',n_jobs=-1)
```

```python
sel_features = selector.fit_transform(Xnew,Y)
```

```python
imp_cols = selector.get_feature_names_out()
```

```python
imp_cols
```

```python
len(imp_cols)
```

## Final Model building

```python
X.head()
```

```python
X_sel = X[imp_cols]
X_sel.head()
```

## Preprocess Categorical features with One Hot encoding

```python
cat_sel,con_sel = catconsep(X_sel)
```

```python
cat_sel
```

```python
con_sel
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
X_cat_sel_enc = ohe.fit_transform(X[cat_sel]).toarray()
X_cat_sel_enc = pd.DataFrame(X_cat_sel_enc,columns=ohe.get_feature_names_out())
```

```python
X_cat_sel_enc.head()
```

## Apply standard scaler on continuous features

```python
from sklearn.preprocessing import StandardScaler
scaler1 = StandardScaler()
X_con_sel_scl = pd.DataFrame(scaler1.fit_transform(X[con_sel]),columns=con_sel)
X_con_sel_scl.head()
```

## Join Cat and Con features

```python
Xnew_sel = X_con_sel_scl.join(X_cat_sel_enc)
```

```python
Xnew_sel.head()
```

```python
Y.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(Xnew_sel,Y, test_size=0.2,random_state=21)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtest.shape
```

```python
ytest.shape
```

## Final Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(xtrain,ytrain)
```

## Evaluate Final Model

```python
ypred_tr = model1.predict(xtrain)
ypred_ts = model1.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
tr_mse = mean_squared_error(ytrain, ypred_tr)
tr_rmse = tr_mse**(1/2)
tr_mae = mean_absolute_error(ytrain,ypred_tr)
tr_r2 = r2_score(ytrain,ypred_tr)
```

```python
print(f'Training MSE : {tr_mse:.2f}')
print(f'Training RMSE : {tr_rmse:.2f}')
print(f'Training MAE : {tr_mae:.2f}')
print(f'Training R2 : {tr_r2:.4f}')
```

```python
ts_mse = mean_squared_error(ytest,ypred_ts)
ts_rmse = ts_mse**(1/2)
ts_mae = mean_absolute_error(ytest,ypred_ts)
ts_r2 = r2_score(ytest, ypred_ts)
```

```python
print(f'Tesing MSE : {ts_mse:.2f}')
print(f'Testing RMSE : {ts_rmse:.2f}')
print(f'Testing MAE : {ts_mae:.2f}')
print(f'Testing R2 : {ts_r2:.4f}')
```
